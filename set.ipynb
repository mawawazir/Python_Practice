```python
set1={"apple","banana","cherry"}
set2={1,2,3}
set3=set1.union(set2)
print(set3)
```

```python
#Add method
set={"apple","bannana","cherry"}
set.add("mango")
print(set)
```

```python
# Difference method
set1={"apple","bannana","cherry"}
set2={"google","microsoft","apple"}
print(set1.difference(set2))
```

```python
# intersection method
set1={"apple","bannana","cherry"}
set2={"google","microsoft","apple"}
print(set1.intersection(set2))
```
