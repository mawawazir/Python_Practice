```python
#Dictionary :-
# A collection of key value-pairs.

#Create dictionary: Dictionaries are written with curly brackets &  have ky and values.
thisdict={
    "Brand":"Ford",
    "Model":"Mustang",
    "year":1984
}
print(thisdict["Model"])

#Properties:-

#1:- ordered:-
              #Dictionaries don't support index-based access.
print(thisdict["Brand"])


#2:- Mutable:-
              #Means we can change the dictionary after creating it.



#3:- Keys must be unique:-
              #A dictionary cannot have two enteries with the same key.
              #Don't allow duplicates.
print(thisdict)

# Lenght of dictionary:-
print(len(thisdict))

# Dictionary data type:
# the values in dictionary items can be of ay data type.

thisdict1 = {
  "brand": "Ford",
  "electric": False,
  "year": 1964,
  "colors": ["red", "white", "blue"]
}

print(thisdict1)

#type():
print(type(thisdict1))

```

```python
#Access items:-
thisdict={
    "Brand":"Ford",
    "Model":"Mustang",
    "year":1984
}
print(thisdict["Brand"])

#.get()
x=thisdict.get("Model",)
print(x)
#keys()
print(thisdict.keys())

#values()

print(thisdict.values())

#items()
print(thisdict.items())
```

```python
#change values:-
thisdict={
    "Brand":"Ford",
    "Model":"Mustang",
    "year":1984
}

thisdict["Brand"]="corolla"
print(thisdict)

thisdict.update({"year":2010})
```

```python
#Add items:-
thisdict={
    "Brand":"Ford",
    "Model":"Mustang",
    "year":1984
}
thisdict["color"]="red"
print(thisdict)

thisdict.update({"name":"Gul_khan_"})
print(thisdict)
```

```python
#Remove dict items:-
thisdict={
    "Brand":"Ford",
    "Model":"Mustang",
    "year":1984,
    "color":"red"
}
thisdict.pop("Model")
print(thisdict)

thisdict.popitem()
print(thisdict)

#del():-
del thisdict["Brand"]
print(thisdict)

#clear():-
thisdict.clear()
print(thisdict)


#items:-


```
