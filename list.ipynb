```python
# List Practice
#lists are:
#ordered
#changeable(mutable)
# indexed
#list items are created using square brackets.

# list can store multiple distinct items in a single variable.
#e.g:
list=["apple",1,"banana",2,"cherry",3]
print(list)
print(list[0])
print(list[1])
```

```python
#Access lit items:
#we can access list items by referring to index number.

list=["apple","banana","mango","orange","watermelon","kiwi"]
print(list[3])
print(list[0:3])

print(list[-1])
print(list[-2])

print(list[-4:-1])
```

```python
#change list items:
list=["apple","banana","cherry","mango","watermelon","kiwi","orange"]
list[0]="melon"
print(list)

list[1]="blackcurrant"
print(list)

```
